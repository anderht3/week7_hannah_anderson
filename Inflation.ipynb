As you can see from these figures, the Hubble parameter seems to be decreasing rapidly during this time.
```python
print(inflation.sol.t[inflation.final])
```

Now we can begin to look at the scale factor a.

```python
plt.plot(inflation.aInfl,inflation.hInfl)
plt.yscale("log")
plt.xlabel("a")
plt.ylabel("H (km\s\Mpc)")
```

Now it is much more clear from this figure that H remains largely constant during this epoch which just a decrease at the end of the inflation epoch (which is what we expect). Now we will plot the slow roll parameter in comparison to the comoving hubble distance as well as the hubble parameter against the comoving hubble distance.

```python
plt.plot(3e8/(inflation.aInfl*inflation.hInfl), inflation.eSlowInfl)
```

```python
plt.plot((3e8/inflation.aInfl*inflation.hInfl), inflation.hInfl/3e8)
```

Here again we can clearly see that the hubble paramtere remains constant over time. We can also see the fluctuations of the $\epsilon_{sr}$ parameter. It is clear that the slow roll parameter is very small during inflation.

Next we will look at a few cosmological parameters that are measured from the CMB: $A_s$ and $n_s$. The measurements that are given depict that $\epsilon_{sr}$ again is very small since $4\epsilon_{sr}$ is subtracted from 1 giving us 0.9682 which shows agreement with our figure. This also means that H is relatively small on the scale of 10^-4 which is what we see here as well.
